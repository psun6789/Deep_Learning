```python
import numpy as np
import h5py

import tensorflow as tf

from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import matplotlib.pyplot as plt
```

```python
# function to Loading the Data from Drive
def loadDataH5(): 
  with h5py.File('earth_data.h5','r') as hf:
    trainX = np.array(hf.get('trainX'))
    trainY = np.array(hf.get('trainY'))
    valX = np.array(hf.get('valX'))
    valY = np.array(hf.get('valY'))
    print (trainX.shape,trainY.shape)
    print (valX.shape,valY.shape)

    return trainX, trainY, valX, valY

trainX, trainY, valX, valY = loadDataH5()
```

```python
trainX = trainX.astype('float32')/255.0
# trainY = trainY.astype('float32')/255.0

valX = valX.astype('float32')/255.0
# valY = valY.astype('float32')/255.0
```

Base Line Model

```python
# Base model with one Convolution Layer

def baseline_model(class_label, trainX, trainY, valX, valY):
  # Convolution -> Flatten -> FC --> softmax_layer
  input_s = trainX.shape[1:]

  base_model = Sequential()

  # model.add(Conv2D(32, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  base_model.add(Conv2D(16, (3,3), input_shape=input_s, activation = 'relu', padding= 'same'))

  base_model.add(MaxPooling2D(2, 2))

  base_model.add(Flatten())

  base_model.add(Dense(32, activation = 'relu') )

  base_model.add(Dense(9, activation = 'softmax') )

  return base_model
```

**Architecture - 1**

```python
# Architecture 1

def  CNN_configurations_1(class_label, trainX, trainY, valX, valY):
  input_s = trainX.shape[1:]

  # CNN1-> MxP1 -> CNN2 -> MxP2 -> Softmax
  architecture_1_model = Sequential()

  # 1st Layer
  architecture_1_model.add(Conv2D(16, (3,3), input_shape=input_s, activation = 'relu', padding= 'same'))
  architecture_1_model.add(MaxPooling2D(2, 2))

  # 2nd Layer
  architecture_1_model.add(Conv2D(32, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_1_model.add(MaxPooling2D(2, 2))

  architecture_1_model.add(Flatten())
  architecture_1_model.add(Dense(64, activation = 'relu') )
  architecture_1_model.add(Dense(9, activation = 'softmax') )

  return architecture_1_model


```

**Architect-2**

```python
def  CNN_configurations_2(class_label, trainX, trainY, valX, valY):
  input_s = trainX.shape[1:]

  # CNN1-> MxP1 -> CNN2 -> MxP2 -> CNN3 -> MxP3 -> Softmax
  architecture_2_model = Sequential()

  # 1st Layer
  architecture_2_model.add(Conv2D(16, (3,3), input_shape=input_s, activation = 'relu', padding= 'same'))
  architecture_2_model.add(MaxPooling2D(2, 2))

  # 2nd Layer
  architecture_2_model.add(Conv2D(32, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_2_model.add(MaxPooling2D(2, 2))

  # 3rd Layer
  architecture_2_model.add(Conv2D(64, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_2_model.add(MaxPooling2D(2, 2))

  architecture_2_model.add(Flatten())
  architecture_2_model.add(Dense(64, activation = 'relu') )
  architecture_2_model.add(Dense(9, activation = 'softmax') )

  return architecture_2_model
```

**ARchitect-3**

```python
def  CNN_configurations_3(class_label, trainX, trainY, valX, valY):
  input_s = trainX.shape[1:]

  # CNN1-> MxP1 -> CNN2 -> MxP2 -> CNN3 -> MxP3 -> CNN4 -> MxP4 -> Softmax
  architecture_3_model = Sequential()

  # 1st Layer
  architecture_3_model.add(Conv2D(16, (3,3), input_shape=input_s, activation = 'relu', padding= 'same'))
  architecture_3_model.add(MaxPooling2D(2, 2))

  # 2nd Layer
  architecture_3_model.add(Conv2D(32, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_3_model.add(MaxPooling2D(2, 2))

  # 3rd Layer
  architecture_3_model.add(Conv2D(64, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_3_model.add(MaxPooling2D(2, 2))

  # 4th Layer
  architecture_3_model.add(Conv2D(128, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_3_model.add(MaxPooling2D(2, 2))

  architecture_3_model.add(Flatten())
  architecture_3_model.add(Dense(64, activation = 'relu') )
  architecture_3_model.add(Dense(9, activation = 'softmax') )

  return architecture_3_model
```

**ARchitecture 4**

```python
def  CNN_configurations_4(class_label, trainX, trainY, valX, valY):
  input_s = trainX.shape[1:]

  # CNN1-> MxP1 -> CNN2 -> MxP2 -> CNN3 -> MxP3 -> CNN4 -> MxP4 -> CNN5 -> MxP5 -> Softmax
  architecture_4_model = Sequential()

  # 1st Layer
  architecture_4_model.add(Conv2D(16, (3,3), input_shape=input_s, activation = 'relu', padding= 'same'))
  architecture_4_model.add(MaxPooling2D(2, 2))

  # 2nd Layer
  architecture_4_model.add(Conv2D(32, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_4_model.add(Conv2D(48, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_4_model.add(MaxPooling2D(2, 2))

  # 3rd Layer
  architecture_4_model.add(Conv2D(64, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_4_model.add(Conv2D(96, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_4_model.add(MaxPooling2D(2, 2))

  # 4th Layer
  architecture_4_model.add(Conv2D(128, (3,3), input_shape=(64,64,3), activation = 'relu', padding= 'same'))
  architecture_4_model.add(MaxPooling2D(2, 2))

  architecture_4_model.add(Flatten())
  architecture_4_model.add(Dense(64, activation = 'relu') )
  architecture_4_model.add(Dense(9, activation = 'softmax') )

  return architecture_4_model
```

```python
def plot_history(history, architecture, details, display):
  plt.style.use('ggplot')
  plt.figure()
  plt.plot(np.arange(0, epochs), history.history['loss'], label='Training Loss')
  plt.plot(np.arange(0, epochs), history.history['val_loss'], label='Validation Loss')
  plt.plot(np.arange(0, epochs), history.history['acc'], label='Training Accuracy')
  plt.plot(np.arange(0, epochs), history.history['val_acc'], label='Validation Accuracy')

  plt.title(architecture + display + "\n" + details)

  plt.xlabel("Epochs --> ")
  plt.ylabel("Loss / Accuracy")

  plt.legend()

  plt.savefig(architecture+display+".png")
```

```python
import time
def run_model(model_no, class_label, epochs, Batch, data_augm, config):
  start = time.time()

  if model_no == 0:
    model = baseline_model(class_label, trainX, trainY, valX, valY)
  elif model_no == 1:
    model = CNN_configurations_1(class_label, trainX, trainY, valX, valY)
  elif model_no == 2:
    model = CNN_configurations_2(class_label, trainX, trainY, valX, valY)
  elif model_no == 3:
    model = CNN_configurations_3(class_label, trainX, trainY, valX, valY)
  elif model_no == 4:
    model = CNN_configurations_4(class_label, trainX, trainY, valX, valY)

  summary = model.summary()
  print("Model Summary", summary)

  model.compile(optimizer = Adam(lr = 0.01), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

  if data_augm == 0:
    history = model.fit(x=trainX, y=trainY, batch_size = Batch, epochs = epochs, validation_data = (valX, valY))
  
  end = time.time()

  return history


```

```python
import time
from tensorflow.keras.preprocessing.image import ImageDataGenerator


def run_model(model_no, class_label, epochs, Batch, data_augm, config):
  start = time.time()

  if model_no == 3:
    model = CNN_configurations_3(class_label, trainX, trainY, valX, valY)

  summary = model.summary()
  print("Model Summary", summary)

  model.compile(optimizer = Adam(lr = 0.01), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

  if data_augm == 1:
    if config == 1:
      train_data = ImageDataGenerator(featurewise_center=False, samplewise_center=False, featurewise_std_normalization=False, samplewise_std_normalization=False, zca_whitening=False, zca_epsilon=1e-06,
              rotation_range=10, width_shift_range=0.1, height_shift_range=0.1,
              brightness_range=None, shear_range=0.0, zoom_range=0.0, channel_shift_range=0.0, fill_mode='nearest', cval=0.0, horizontal_flip=False, vertical_flip=False, rescale=None, preprocessing_function=None,
              data_format=None, validation_split=0.0, dtype=None)
    elif config == 2:
      train_data = ImageDataGenerator(featurewise_center=False, samplewise_center=False, featurewise_std_normalization=False, samplewise_std_normalization=False, zca_whitening=False, zca_epsilon=1e-06,
              rotation_range=45, width_shift_range=0.33, height_shift_range=0.33, zoom_range=0.33, horizontal_flip=True, vertical_flip=True,
              brightness_range=None, shear_range=0.0, channel_shift_range=0.0, fill_mode='nearest', cval=0.0,  rescale=None, preprocessing_function=None,
              data_format=None, validation_split=0.0, dtype=None)
    elif config == 3:
      train_data = ImageDataGenerator(featurewise_center=False, samplewise_center=False, featurewise_std_normalization=False, samplewise_std_normalization=False, zca_whitening=False, zca_epsilon=1e-06,
              rotation_range=90, width_shift_range=0.50, height_shift_range=0.50, zoom_range=1.0, horizontal_flip=True, vertical_flip=True, shear_range=0.5,
              brightness_range=None,  channel_shift_range=0.0, fill_mode='nearest', cval=0.0,  rescale=None, preprocessing_function=None,
              data_format=None, validation_split=0.0, dtype=None)
    else:
      train_data = ImageDataGenerator(featurewise_center=False, samplewise_center=False, featurewise_std_normalization=False, samplewise_std_normalization=False, zca_whitening=False, zca_epsilon=1e-06,
              rotation_range=45, width_shift_range=[0.10, 0.50, 0.75], height_shift_range=[0.10, 0.50, 0.75], zoom_range=0.5, horizontal_flip=True, vertical_flip=True, shear_range=0.2,
              brightness_range=None,  channel_shift_range=0.0, fill_mode='nearest', cval=0.0,  rescale=None, preprocessing_function=None,
              data_format=None, validation_split=0.0, dtype=None)


    train_data.fit(trainX)
    history = model.fit(
        train_data.flow(trainX, trainY, batch_size = Batch),
        steps_per_epoch = 1000 // batch_size, epochs = epochs, validation_data = (valX, valY)
        )
  end = time.time()

  return history


```
